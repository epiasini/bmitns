```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(constrained_layout=True, figsize=(5,3))


texture_values = np.arange(start=0.02, stop=1.01, step=0.07)

ax.stem(0, 1, linefmt='#008000')
markerline, stemlines, baseline = ax.stem(texture_values, np.full_like(texture_values, 1/texture_values.size), linefmt='#ffa500')
baseline.set_visible(False)
ax.axhline(0, color='k', lw=0.5)
ax.set_xlabel('Pattern intensity (stimulus) s')
ax.set_ylabel('Class cond. stim. distribution p(s|c)')

ax.annotate('p(s|no texture)', (0.02, 0.8), color='#008000')
ax.annotate('p(s|texture)', (0.6, 0.15), color='#ffa500')

fig.savefig('figures/caramellino_ccsd.png', dpi=300)
```
